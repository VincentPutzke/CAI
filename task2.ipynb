```python
# Importing WordCloud
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
# Importing and mounting the drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Loading the sample text from the file
with open("/content/drive/MyDrive/CAI/JaneEyre.txt", "r") as file:
  text = file.read()
```

```python
# Building the cloud: white bg, 10 min font size
wordcloud = WordCloud(
  width=800,
  height=400,
  background_color="white",
  stopwords=set(STOPWORDS),
  min_font_size=10,
).generate(text)

# Plotting using plt
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
# Building the cloud: black bg, 0.1 min font size, custom stopwords
wordcloud = WordCloud(
  width=800,
  height=400,
  background_color="black",
  stopwords=set(["the", "a", "an"]),
  min_font_size=0.1,
).generate(text)

# Plotting using plt
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
# Building the cloud: turquoise bg, 0.1 min font size, no stop words
wordcloud = WordCloud(
  width=1600,
  height=800,
  background_color="turquoise",
  stopwords=[],
  min_font_size=0.1,
).generate(text)

# Plotting using plt
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
