```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

# An interesting text about proboscidea, the scientific name for
# elephants and their ancestors, a hobby of mine.
text = "Proboscidea, the order of mammals that includes modern elephants, has a rich evolutionary history spanning over 60 million years. Among the diverse proboscideans that once roamed the Earth were the tapir-like Moeritherium, the shovel-tusked Platybelodon, the four-tusked Gomphotherium, the woolly mammoth, and the American mastodon. These ancient proboscideans exhibited a wide range of adaptations, from the semi-aquatic lifestyle of Moeritherium to the grazing habits of later species, reflecting their ability to thrive in various environments across Africa, Eurasia, and the Americas. The evolution of proboscideans saw remarkable changes in their anatomy, including increases in body size, elongation of limbs, expansion of the skull, and most notably, the development of their iconic trunks and tusks."

# Tokenize the text
tokens = word_tokenize(text)
# Print the tokens
print(tokens)
print("-----------------------")
# Print the count
print(f"Count: {len(tokens)}")
print("-----------------------")

# Calculate the frequencies
fdist = nltk.FreqDist()
for word in tokens:
  fdist[word.lower()] += 1

# Print the tokens sorted by their frequency with their respective occurrences.
for item in fdist.most_common(len(tokens)):
  print(item[0] + ": " + str(item[1]))
```
